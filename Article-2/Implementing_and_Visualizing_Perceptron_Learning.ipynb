```python
import numpy as np
import matplotlib.pyplot as plt

# Inputs (features) and labels
x = np.array([3, 5, 6, 8, 1])
y = np.array([1, 1, -1, -1, 1])

# Initialize weights and bias
w = 0
b = 0

# Perceptron training loop
for _ in range(25):
    y_hat = np.sign(w * x + b)  # Predict output using current weights and bias
    correct = y == y_hat  # Check if predictions are correct

    # Update weights and bias for each misclassified point
    for i in range(len(y)):
        w += (y[i] - y_hat[i]) * x[i]
        b += (y[i] - y_hat[i])

    # Output which predictions are correct at each iteration
    print(correct)

```

```python
# Final weights and bias after training
print(f'Final weight (w): {w}')
print(f'Final bias (b): {b}')

# Plot the data points
plt.scatter(x[y == 1], y[y == 1], color='blue', label='Class 1 (y=1)')
plt.scatter(x[y == -1], y[y == -1], color='red', label='Class -1 (y=-1)')

# Plot decision boundary
# The decision boundary is where w * x + b = 0
# Rearrange this to get x = -b / w
x_values = np.linspace(min(x) - 1, max(x) + 1, 100)
decision_boundary = -(b / w) * np.ones_like(x_values)

plt.plot(x_values, decision_boundary, color='green', linestyle='--', label='Decision Boundary')

# Labels and title
plt.axhline(0, color='black', linewidth=0.5)  # Line at y = 0
plt.axvline(0, color='black', linewidth=0.5)  # Line at x = 0
plt.xlabel('Input Feature (x)')
plt.ylabel('Decision Boundary (y)')
plt.title('Perceptron Decision Boundary')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Inputs (features) and labels
x = np.array([3, 1, 7, 8, 4])
y = np.array([1, 1, -1, -1, 1])

# Initialize weights and bias
w = 0
b = 0
learning_rate = 0.01  # Learning rate to control updates

# Perceptron training loop
for _ in range(50):  # More iterations for better convergence
    y_hat = np.sign(w * x + b)  # Predict output using current weights and bias
    correct = y == y_hat  # Check if predictions are correct

    # Update weights and bias for each misclassified point
    for i in range(len(y)):
        if y[i] != y_hat[i]:  # Update only for misclassified points
            w += learning_rate * (y[i] - y_hat[i]) * x[i]
            b += learning_rate * (y[i] - y_hat[i])

    # Output which predictions are correct at each iteration
    print(f"Iteration {_ + 1}: Correctly classified: {np.sum(correct)} / {len(y)}")

# Final weights and bias after training
print(f'Final weight (w): {w}')
print(f'Final bias (b): {b}')

# Plot the data points
plt.scatter(x[y == 1], y[y == 1], color='blue', label='Class 1 (y=1)')
plt.scatter(x[y == -1], y[y == -1], color='red', label='Class -1 (y=-1)')

# Plot decision boundary
# The decision boundary is where w * x + b = 0
# Rearrange this to get x = -b / w
x_values = np.linspace(min(x) - 1, max(x) + 1, 100)
decision_boundary = -(b + w * x_values) / w

plt.plot(x_values, decision_boundary, color='green', linestyle='--', label='Decision Boundary')

# Labels and title
plt.axhline(0, color='black', linewidth=0.5)  # Line at y = 0
plt.axvline(0, color='black', linewidth=0.5)  # Line at x = 0
plt.xlabel('Input Feature (x)')
plt.ylabel('Decision Boundary (y)')
plt.title('Perceptron Decision Boundary')
plt.legend()
plt.grid(True)
plt.show()
```
