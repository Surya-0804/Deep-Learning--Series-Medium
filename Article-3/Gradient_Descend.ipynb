```python
import numpy as np

# Example data (input x and target y)
x = np.array([1, 2, 3, 4])
y = np.array([2, 4, 6, 8])

# Initialize weights and learning rate
w = 0
learning_rate = 0.01
epochs = 100

# Gradient Descent loop
for epoch in range(epochs):
    # Predict output
    y_pred = w * x

    # Calculate Mean Squared Error (MSE)
    loss = np.mean((y - y_pred)**2)

    # Calculate the gradient
    gradient = -2 * np.mean(x * (y - y_pred))

    # Update weights
    w = w - learning_rate * gradient

    # Print loss at each epoch
    print(f'Epoch {epoch + 1}: Loss = {loss:.4f}, Weight = {w:.4f}')

print(f'Final Weight: {w}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data (input x and target y)
x = np.array([1, 2, 3, 4])
y = np.array([2, 4, 6, 8])

# Initialize weights and learning rate
w = 0
learning_rate = 0.01
epochs = 100

# Store loss values for visualization
loss_history = []

# Gradient Descent loop
for epoch in range(epochs):
    # Predict output
    y_pred = w * x

    # Calculate Mean Squared Error (MSE)
    loss = np.mean((y - y_pred)**2)

    # Store the loss for plotting
    loss_history.append(loss)

    # Calculate the gradient
    gradient = -2 * np.mean(x * (y - y_pred))

    # Update weights
    w = w - learning_rate * gradient

    # Optionally, print the loss and weights at every epoch
    print(f'Epoch {epoch + 1}: Loss = {loss:.4f}, Weight = {w:.4f}')

# Plot the loss over epochs
plt.figure(figsize=(12, 6))

# Plot the loss history
plt.subplot(1, 2, 1)
plt.plot(range(1, epochs + 1), loss_history, label="Loss")
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss over Epochs')
plt.grid(True)

# Plot the fitted line at the final stage
plt.subplot(1, 2, 2)
plt.scatter(x, y, color='blue', label='True data')
plt.plot(x, w * x, color='red', label=f'Fitted line (w={w:.4f})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Fitted Line after Gradient Descent')
plt.legend()
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()

print(f'Final Weight: {w}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function f(x) = x^2
def func(x):
    return x ** 2

# Derivative f'(x) = 2x
def derivative(x):
    return 2 * x

# Values
x = np.linspace(-10,10,100)

# Create plot
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(x, func(x), label='f(x) = x^2')
ax.plot(x, derivative(x), label="f'(x) = 2x")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
plt.grid(True)
plt.show()
```

```python
# Gradient Descent Algorithm
def gradient_descent(start_x, derivative, learning_rate, epochs):
    x = start_x
    history = [x]

    for _ in range(epochs):
        grad = derivative(x)
        x -= learning_rate * grad
        history.append(x)

    return history

# Parameters
start_x = -7.5
learning_rate = 0.1
epochs = 20

# Run Gradient Descent
history = gradient_descent(start_x, derivative, learning_rate, epochs)

# Plot
fig, ax = plt.subplots(figsize=(8,6))
x = np.linspace(-8, 8, 400)
ax.plot(x, func(x), label='f(x) = x^2')
ax.scatter(
  history,
  [func(i) for i in history],
  color='red',
  label='Gradient Descent'
)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
plt.grid(True)
plt.show()
```

```python
# Different learning rates
learning_rates = [0.01, 0.1, 0.9]

# Run Gradient Descent and plot
fig, ax = plt.subplots(figsize=(8,6))
x = np.linspace(-8, 8, 400)

colors = ['purple', 'green', 'red']
ax.plot(x, func(x), label='f(x) = x^2')

for lr, color in zip(learning_rates, colors):
    history = gradient_descent(start_x, derivative, lr, epochs)
    ax.scatter(history, [func(i) for i in history], color=color, label=f'Learning rate: {lr}')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
plt.grid(True)
plt.show()
```

# **The following code visualizes the steps of Gradient Descent on the function f(x)=x.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Function f(x) = x^2
def func(x):
    return x ** 2

# Derivative f'(x) = 2x
def derivative(x):
    return 2 * x

# Gradient Descent Algorithm
def gradient_descent(start_x, learning_rate, epochs):
    x = start_x
    history = [x]
    for _ in range(epochs):
        x -= learning_rate * derivative(x)
        history.append(x)
    return history

# Plot the function and the gradient descent steps
def plot_gradient_descent(history, learning_rate):
    x = np.linspace(-8, 8, 100)
    plt.plot(x, func(x), label='f(x) = x^2')
    plt.scatter(history, [func(i) for i in history], color='red', label='Gradient Descent Steps')
    plt.title(f'Gradient Descent with Learning Rate {learning_rate}')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Run gradient descent and visualize
start_x = -7.5
learning_rate = 0.1
epochs = 20
history = gradient_descent(start_x, learning_rate, epochs)

plot_gradient_descent(history, learning_rate)

# Test different learning rates
learning_rates = [0.01, 0.1, 0.5]
for lr in learning_rates:
    history = gradient_descent(start_x, lr, epochs)
    plot_gradient_descent(history, lr)
```
