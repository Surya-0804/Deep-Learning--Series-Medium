```python
import numpy as np
import matplotlib.pyplot as plt

# Generate dummy data for loss values across epochs
epochs = np.arange(1, 21)

# Simulate MSE and Cross Entropy Loss for demonstration
mse_loss = np.random.uniform(0.5, 0.05, size=len(epochs))  # Simulated decreasing MSE loss
cross_entropy_loss = np.random.uniform(0.7, 0.1, size=len(epochs))  # Simulated decreasing cross-entropy loss

# Plot MSE and Cross Entropy Loss
plt.figure(figsize=(10, 6))
plt.plot(epochs, mse_loss, label="MSE Loss", marker='o', linestyle='--', color='b')
plt.plot(epochs, cross_entropy_loss, label="Cross Entropy Loss", marker='s', linestyle='-', color='r')

# Adding title and labels
plt.title('MSE Loss vs Cross Entropy Loss over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```
